```python
from PIL import Image, ImageDraw, ImageFont


# Function to calculate dynamic height and table width based on data
def calculate_image_height(data):
    base_height = 250  # Starting height to fit header info
    table_height = 250  # Height for each table (1 table per term)
    total_height = base_height + table_height * len(data["Terms"])
    return total_height

# Function to draw a table for each term
def draw_table(draw, term_data, x_start, y_start, width, font):
    table_height = 200
    header_height = 40
    row_height = 30

    # Draw table headers
    draw.rectangle([x_start, y_start, x_start + width, y_start + header_height], fill="lightgray")
    draw.text((x_start + 10, y_start + 10), "Subject", fill="black", font=font)
    draw.text((x_start + 200, y_start + 10), "Score", fill="black", font=font)
    draw.text((x_start + 300, y_start + 10), "Attendance", fill="black", font=font)
    draw.text((x_start + 400, y_start + 10), "Remarks", fill="black", font=font)

    # Draw table rows for subjects
    y_position = y_start + header_height
    for subject in term_data["Subjects"]:
        draw.text((x_start + 10, y_position + 5), subject["Subject"], fill="black", font=font)
        draw.text((x_start + 200, y_position + 5), f"{subject['Score']}", fill="black", font=font)
        draw.text((x_start + 300, y_position + 5), subject["Attendance"], fill="black", font=font)
        draw.text((x_start + 400, y_position + 5), subject["Remarks"], fill="black", font=font)
        y_position += row_height

    # Draw GPA and overall remarks for the term
    draw.rectangle([x_start, y_position, x_start + width, y_position + 40], fill="lightgray")
    draw.text((x_start + 10, y_position + 10), f"Total: {term_data['Total']}", fill="black", font=font)
    y_position += 40
    draw.text((x_start + 10, y_position + 5), f"Overall Remarks: {term_data['Term_Remarks']}", fill="black", font=font)

    return y_position + 50  # Return the updated y_position after the table

# Function to create the report card image with multiple terms in tables
def generate_report_card_image(data, filename="report_card_class9.jpg"):
    width = 800
    height = calculate_image_height(data)  # Dynamic height based on number of terms
    image = Image.new('RGB', (width, height), "white")
    draw = ImageDraw.Draw(image)

    # Load a larger font size (20)
    try:
        font = ImageFont.truetype("arial.ttf", 40)
    except IOError:
        font = ImageFont.load_default(15)  # Use default font if truetype font is not available

    # Draw student details side by side
    draw.text((20, 20), f"Name: {data['Name']}", fill="black", font=font)
    draw.text((200, 20), f"Age: {data['Age']}", fill="black", font=font)
    draw.text((400, 20), f"Class: {data['class']}", fill="black", font=font)
    draw.text((600, 20), f"Student ID: {data['Student_ID']}", fill="black", font=font)

    # Position for the first term table
    y_position = 50

    for term in data["Terms"]:
        draw.text((20, y_position), f"Term {term['Term']}", fill="black", font=font)
        y_position += 30
        y_position = draw_table(draw, term, 20, y_position, width - 40, font)  # Draw each term's table

        # Add space between tables
        y_position += 10

    # Save the image
    image.save(filename)
    print(f"Report card saved as {filename}")

# Generate the report card image
generate_report_card_image(student_data)
```
